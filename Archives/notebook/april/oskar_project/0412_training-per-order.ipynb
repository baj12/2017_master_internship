```python
import pandas as pd
```

```python
df = pd.read_csv('/home/savy/Documents/2017_master_internship/GenePredictionTask/Genomes/genomeDownloader/GCA_GCF_insectGenomes.csv')
```

```python
GCFDict = {'species':[], 'genome_id':[], 'order':[], 'family':[]}

for index in df.index :
    if 'GCF' in df['genome_id'][index]:
        GCFDict['species'].append(df['sp_name'][index])
        GCFDict['genome_id'].append(df['genome_id'][index])
        GCFDict['order'].append(df['order_name'][index])
        GCFDict['family'].append(df['family_name'][index])
```

```python
GCF_df = pd.DataFrame(GCFDict)
GCF_df.groupby(['order','family']).count()
```

```python
GCADict = {'species':[], 'genome_id':[], 'order':[], 'family':[]}

for index in df.index :
    if 'GCA' in df['genome_id'][index]:
        GCADict['species'].append(df['sp_name'][index])
        GCADict['genome_id'].append(df['genome_id'][index])
        GCADict['order'].append(df['order_name'][index])
        GCADict['family'].append(df['family_name'][index])
```

```python
GCA_df = pd.DataFrame(GCADict)
GCA_df.groupby(['order','family']).count()
```
